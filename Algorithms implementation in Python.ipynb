## 2. Plotting
To plot in Python we use matplotlib. Install it by running:

    pip install matplotlib

And import it

```python
from pylab import *
```

Here is how to plot some values. The function plot accepts two arguments: the x's and the y's. Here we have only values for `y` so we use x = [0, 1, 2, 3, 4, 5, 6].

```python
y = [5, -3, 2, 0, 8, 1]
x = range(6)
plot(x, y, 'o-')
show()
```

Plotting bars:

```python
bar(x, y)
show()
```

Plotting points:

```python
plot(x, y, 'o')
show()
```

Plotting several series in the same plot

```python
for i in range(5):
    x = range(6)
    y = [randrange(1, 10) for j in range(6)]
    plot(x, y, 'o-')
show()
```

Plotting in log scale. Look at the scale of the plot to understand what it means.

```python
subplot(111, xscale="log", yscale="log")

x = [2**p for p in range(1, 20)]
y = [2**p for p in range(1, 20)]
    
plot(x, y, 'o-')
show()
```
